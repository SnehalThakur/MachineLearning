# Linear regression with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data
area = [1.2,2.4,5.6,8.7,5.9,6.8,5.4,2.9,3.6,5.8,1.8,2.5,8.4,5.6,6.2]   # feature - x
price =[190,280,480,760,680,720,570,220,420,490,200,200,890,680,720]   # label - y^
len(area),len(price)
```

```python
range(5)
```

```python
x = np.array(area)
y = np.array(price)
m = 10
c = 20
LR = 0.002
for i in range(5):
    # get the predictions
    yhat = m*x + c
    
    # calculate Mean Squared Error (MSE)
    error = np.mean(np.square(y-yhat))     # MSE = 1\n SUM(y - y^)**2
    
    # Gradient descent
    dedm = 2*(yhat-y).dot(x)                   # 2*(y^ -y).x
    dedc = 2*(yhat-y).dot(np.ones(len(x)))     # 2*(y^ -y).x
    
    m = m - LR*dedm       # m(new) = m(old) - learning rate * derivative of Error w.r.t m(old)
    c = c - LR*dedc       # c(new) = c(old) - learning rate * derivative of Error w.r.t c(old)
    iteration = i+1
    print("Iteration ", iteration)
    print("error=", error)
    print("m(new)=", m)
    print("c(new)=", c)
    print("-------------------------")
    
```

```python
x = np.array(area)
y = np.array(price)
m = 10
c = 20

LR = 0.002

for i in range(20):
    # get the predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y-yhat))
    
    # Gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(error)
    
    plt.scatter(x,y)
    plt.plot(x,yhat,c='r')
    plt.show()
```

```python
x = np.array(area)
y = np.array(price)
m = 98
c = 27

LR = 0.002

for i in range(1000):
    # get the predictions
    yhat = m*x + c
    
    # calculate error
    error = np.mean(np.square(y-yhat))
    
    # Gradient descent
    dedm = 2*(yhat-y).dot(x)
    dedc = 2*(yhat-y).dot(np.ones(len(x)))
    
    m = m - LR*dedm
    c = c - LR*dedc
    
    print(error)
```

## Linear Regression with scikit-learn

```python
x = np.array(area)
x
```

```python
x = np.array(area).reshape(15,1)
y = np.array(price)
```

```python
x
```

### Gradient descent way

```python
from sklearn.linear_model import SGDRegressor

model = SGDRegressor(learning_rate='invscaling',tol=0.000001,max_iter=1000,verbose=True)

# train the model with the data
model.fit(x,y)
```

```python
print("values of m and c ",model.coef_,model.intercept_)
```

### OLS approach

```python
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()

# trian the model
model2.fit(x,y)
```

```python
print("values of m and c ",model2.coef_,model2.intercept_)
```

```python
yhat = model.predict(x)
plt.scatter(x,y)
plt.plot(x,yhat,c='r')
plt.show()
```
